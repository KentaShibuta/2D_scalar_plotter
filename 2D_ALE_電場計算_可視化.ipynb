```
!rm ./*.csv # 既にアップロードされているcsvファイルを削除
!pip install pandas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
from matplotlib import ticker, cm, colors

f_name = files.upload()

for fn in f_name.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(f_name[fn])))
  
  f_name = fn

print("read %s" % (f_name))


f_domain_name = files.upload()

# nbool, nbool3, cell_matelの情報
for fn in f_domain_name.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(f_domain_name[fn])))
  
  f_domain_name = fn

print("read %s" % (f_domain_name))

data_domain = np.loadtxt(f_domain_name, delimiter=",", skiprows=0)

elem_num = data_domain[:, 0] # 要素番号
nbool = data_domain[:, 1:5] # nbool
nbool3 = data_domain[:, 5:9] # nbool3
cell_matel = data_domain[:, 9] # cell_matel

nelem = cell_matel.shape[0] # 要素数

print("elem_num = %d" % nelem)

pd_para_data = pd.read_csv(f_name, nrows=1) # パラメータ
data = np.loadtxt(f_name, delimiter=",", skiprows=3) #　節点値

# パラメータの格納
time_data = pd_para_data.iat[0,1]

xelem = pd_para_data.iat[0,2]
yelem = pd_para_data.iat[0,3]
l_elem = pd_para_data.iat[0,4]
g_elem = pd_para_data.iat[0,5]

xnode = xelem + 1
ynode = yelem + 1

nnode = xnode * ynode

# 節点座標と節点値の格納
x = data[:, 1]

y = data[:, 2]
t = data[:, 5]

# 計算格子の可視化
for i in range(int(nelem)):
  plt.plot([x[int(nbool[i][0])], x[int(nbool[i][1])]], [y[int(nbool[i][0])], y[int(nbool[i][1])]], color='k', lw=0.7)
  plt.plot([x[int(nbool[i][1])], x[int(nbool[i][2])]], [y[int(nbool[i][1])], y[int(nbool[i][2])]], color='k', lw=0.7)
  plt.plot([x[int(nbool[i][2])], x[int(nbool[i][3])]], [y[int(nbool[i][2])], y[int(nbool[i][3])]], color='k', lw=0.7)
  plt.plot([x[int(nbool[i][3])], x[int(nbool[i][0])]], [y[int(nbool[i][3])], y[int(nbool[i][0])]], color='k', lw=0.7)

# 界面の可視化
for i in range(int(nelem)):
  for j in range(4):
    if nbool3[i][j] == -1: # 界面
      if j == 3:
        plt.plot([x[int(nbool[i][j])], x[int(nbool[i][0])]], [y[int(nbool[i][j])], y[int(nbool[i][0])]], color='k', lw=5.0)
      else:
        plt.plot([x[int(nbool[i][j])], x[int(nbool[i][j+1])]], [y[int(nbool[i][j])], y[int(nbool[i][j+1])]], color='k', lw=5.0)

    
    elif cell_matel[i] != cell_matel[int(nbool3[i][j])]:
      if j == 3:
        plt.plot([x[int(nbool[i][j])], x[int(nbool[i][0])]], [y[int(nbool[i][j])], y[int(nbool[i][0])]], color='k', lw=5.0)
      else:
        plt.plot([x[int(nbool[i][j])], x[int(nbool[i][j+1])]], [y[int(nbool[i][j])], y[int(nbool[i][j+1])]], color='k', lw=5.0)


# 節点座標と節点値の2次元配列を作成
xx = x.reshape(xnode, ynode)
yy = y.reshape(xnode, ynode)
tt = t.reshape(xnode, ynode)

X = np.array(xx)
Y = np.array(yy)
T = np.array(tt)

plt.contourf(X, Y, T, 100, cmap = 'rainbow')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.colorbar(label='$E_z$')
plt.axes().set_aspect('equal')
plt.savefig('/content/drive/My Drive/sample.svg')
#plt.show()
```
